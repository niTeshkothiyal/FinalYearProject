```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

For 1st CSV file 


```python
file_path = r"C:\Users\91706\Desktop\FinalYearProject\pecan_1_min_part1.csv"  
df = pd.read_csv(file_path, index_col=0)     

print(" Data loaded successfully!")
print(" Columns:", df.columns.tolist())
print(" First few rows:")
display(df.head(3))
print(" Shape:", df.shape)
```

```python
house_ids = sorted(df['dataid'].unique())

print("Total Unique Houses:", len(house_ids))
print("House IDs:")
for hid in house_ids:
    print(hid)
```

```python
# Aggregate to hourly level
df.index = pd.to_datetime(df.index, utc=True, errors='coerce')  
df = df.sort_index()


print("Index type:", type(df.index))
print("Index timezone:", df.index.tz)


hourly = (
    df.groupby('dataid')
      .resample('1h')['use']
      .mean()
      .reset_index()
)

print(" Aggregation successful!")
display(hourly.head())
print(" Shape after aggregation:", hourly.shape)
```

Generating hourly CSV file for 1st CSV 

```python
import datetime
#ts = datetime.datetime.now().strftime("%Y%m%d_%H%M")
hourly.to_csv(f"hourly_data_p_s1.csv", index=False)
print("Hourly CSV saved successfully!")
```
